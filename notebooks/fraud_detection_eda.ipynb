```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("creditcard.csv")

# Display basic info
df.head()
df.info()
df.describe()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
 #Check for Duplicate Rows
df.duplicated().sum()
```

```python
#Removing the Duplicate
df = df.drop_duplicates()
df.shape
```

```python
df.dtypes
```

```python
#Clean & Engineer 'Time' Feature
df["Hour"] = (df["Time"] // 3600) % 24
df["Minute"] = (df["Time"] // 60) % 60
df["Second"] = df["Time"] % 60
```

```python
#Extract hour from the "Time" column
df['Hour'] = (df['Time'] // 3600).astype(int)
```

```python
#Count transactions by hour
hour_counts = df['Hour'].value_counts().sort_index()
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df["Amount_Scaled"] = scaler.fit_transform(df["Amount"].values.reshape(-1,1))
```

```python
df["Class"].value_counts(normalize=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count the number of transactions per class
class_counts = df['Class'].value_counts()

# Prepare the plot
sns.set_style("whitegrid")
plt.figure(figsize=(9,6))

# FIXED: add hue="Class" and legend=False
ax = sns.countplot(
    x="Class", 
    hue="Class",
    data=df, 
    palette=["steelblue", "crimson"],
    legend=False
)

# Add labels on top of bars
for p in ax.patches:
    height = p.get_height()
    x = p.get_x() + p.get_width() / 2 

    ax.text(x, height + 200,
            f"{height:,}",
            ha='center', fontsize=12, fontweight="bold")

# Titles
plt.title("Class Distribution: Fraud vs Non-Fraud Transactions", fontsize=15, fontweight='bold')
plt.xlabel("Class (0 = Normal, 1 = Fraud)", fontsize=12)
plt.ylabel("Number of Transactions", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count the number of transactions per class
class_counts = df['Class'].value_counts().to_dict()

sns.set_style("whitegrid")
plt.figure(figsize=(10,6))

# FIXED: Add hue="Class" to avoid the FutureWarning
ax = sns.boxplot(
    x="Class",
    y="Amount",
    hue="Class",
    data=df,
    palette=["steelblue", "crimson"],
    legend=False
)

# Add count labels above each box
for class_value, count in class_counts.items():
    ax.text(
        class_value,                
        df["Amount"].max() * 1.05,  
        f"Count: {count:,}",
        ha='center',
        fontsize=11,
        fontweight="bold",
        color="black"
    )

plt.title("Transaction Amounts: Fraud vs Non-Fraud", fontsize=12, fontweight='bold')
plt.xlabel("Class (0 = Normal, 1 = Fraud)", fontsize=12)
plt.ylabel("Transaction Amount", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Histogram to see the overall distribution of transaction amounts
sns.histplot(df["Amount"], bins=50, kde=True)
plt.title("Distribution of Transaction Amounts")
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(12,6))
hour_counts.plot(kind='bar')

plt.title("Number of Transactions by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Transactions")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculate fraud rate per hour
hourly_fraud_rate = df.groupby("Hour")["Class"].mean()

# Bar chart of fraud rate by hour
hourly_fraud_rate.plot(kind="bar", figsize=(12,5))
plt.title("Fraud Rate by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Fraud Rate")
plt.show()
```

```python
plt.figure(figsize=(14, 6))  #  Wider plot

sns.kdeplot(df[df["Class"] == 0]["V1"], label="Non-Fraud", fill=True)
sns.kdeplot(df[df["Class"] == 1]["V1"], label="Fraud", fill=True)

plt.title("Distribution of V1: Fraud vs Non-Fraud", fontsize=14)
plt.xlabel("V1", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Credit Card Fraud Detection â€“ Portfolio Notebook

# This notebook contains:
# 1. Project Overview
# 2. Data Loading
# 3. Data Cleaning & Preprocessing
# 4. Exploratory Data Analysis (EDA)
# 5. Insights & Recommendations
# 6. Final Summary

# =============================
# 1. PROJECT OVERVIEW
# =============================

"""
This project analyzes 284,807 credit card transactions to identify the underlying
patterns of fraudulent behavior. The goal is to extract actionable insights
that help detect fraud more effectively.
"""

# =============================
# 2. LOAD DATA
# =============================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("creditcard.csv")  # replace if using another path

df.head()

# =============================
# 3. DATA CLEANING & PREPROCESSING
# =============================

# Check for missing values
df.isnull().sum()

# Basic statistics
df.describe()

# =============================
# 4. EXPLORATORY DATA ANALYSIS (EDA)
# =============================

# ----- A. Class Distribution -----
plt.figure(figsize=(8,5))
class_counts = df['Class'].value_counts()
ax = sns.countplot(x="Class", hue="Class", data=df, palette=["steelblue", "crimson"], legend=False)

# Add exact values on bars
for p in ax.patches:
    count = p.get_height()
    ax.text(p.get_x() + p.get_width()/2, count + 200,
            f"{count:,}", ha='center', fontsize=11, fontweight="bold")

plt.title("Class Distribution: Fraud vs Non-Fraud", fontsize=14)
plt.xlabel("Class (0 = Normal, 1 = Fraud)")
plt.ylabel("Count")
plt.show()

# ----- B. Transaction Amount Distribution (Boxplot) -----
plt.figure(figsize=(10,6))
ax = sns.boxplot(x="Class", y="Amount", hue="Class", data=df,
                 palette=["steelblue", "crimson"], legend=False)

# Show median values
medians = df.groupby("Class")["Amount"].median()

for tick, label in enumerate(ax.get_xticklabels()):
    class_value = int(label.get_text())
    median_val = medians[class_value]
    ax.text(tick, median_val + 10,
            f"Median: {median_val:.2f}", ha='center', fontsize=11)

plt.title("Transaction Amounts: Fraud vs Non-Fraud", fontsize=14)
plt.show()

# ----- C. Hour-of-Day Analysis -----
df['Hour'] = (df['Time'] // 3600) % 24
hour_counts = df.groupby(['Hour', 'Class']).size().unstack(fill_value=0)

plt.figure(figsize=(12,6))
hour_counts.plot(kind='bar', figsize=(12,6))
plt.title("Transactions by Hour: Fraud vs Non-Fraud")
plt.xlabel("Hour of Day")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# ----- D. PCA Feature Distribution (Example: V1) -----
plt.figure(figsize=(12,6))
sns.kdeplot(df[df["Class"]==0]["V1"], label="Non-Fraud", shade=True)
sns.kdeplot(df[df["Class"]==1]["V1"], label="Fraud", shade=True)
plt.title("Distribution of V1: Fraud vs Non-Fraud", fontsize=14)
plt.xlabel("V1")
plt.ylabel("Density")
plt.legend()
plt.show()

# =============================
# 5. INSIGHTS & RECOMMENDATIONS
# =============================

"""
Key Insights:
1. Fraud transactions are extremely rare (0.17%).
2. Fraud amounts tend to be higher and show extreme outliers.
3. Fraud occurs more frequently at night (12 AM â€“ 4 AM).
4. PCA components show clear separation between fraud and non-fraud.

Recommendations:
- Increase monitoring during suspicious hours.
- Add verification for unusually high-value transactions.
- Use class-weighting or SMOTE for model development.
- Deploy anomaly detection for early warning.
"""

# =============================
# 6. FINAL SUMMARY
# =============================

"""
This analysis provides a strong foundation for building a fraud detection model.
Patterns identified in transaction behavior, time patterns, and PCA feature
separation can significantly improve machine learning performance.
"""
```
